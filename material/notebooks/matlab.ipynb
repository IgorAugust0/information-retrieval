Example: exponential

```python
import numpy as np
import matplotlib.pyplot as plt

x2 = np.arange(-5, 5, 0.01)
y2 = np.exp(x2)

plt.plot(x2, y2, "darkblue")
plt.show()
```

```python
# varias plotagens na mesma janela

plt.subplot(211)  # 2 linhas, 1 coluna, 1a figura
plt.plot(x, y, "darkblue")  # linha azul escura

plt.subplot(212)  # 2 linhas, 1 coluna, 2a figura
plt.plot(x2, y2, "--r")  # linha tracejada, vermelha
plt.show()
```

Program that plots the graph of a sin(x) function saving the graph into a figure file

```python
import numpy as np
import matplotlib.pyplot as plt

fig = plt.figure()  # create figure
x3 = np.arange(-np.pi, np.pi, 0.01)
y3 = np.sin(x3)
plt.plot(x3, y3, "darkblue")
plt.show()
fig.savefig("sin.png")  # save figure
```

Example: several simultaneous graph tabs

```python
import numpy as np
import matplotlib.pyplot as plt

x4 = np.arange(-5 * np.pi, 5 * np.pi, 0.01)
y4 = np.cos(x4)

x5 = np.arange(-10, 10, 0.01)
y5 = x5**2 + 2 * x5  # y = x^2 + 2x

plt.figure(1)  # create figure 1
plt.plot(x4, y4, "darkblue")
plt.axis((-5 * np.pi, 5 * np.pi, -5, 5))  # set axis range (x_min, x_max, y_min, y_max
plt.show()

plt.figure(2)  # create figure 2
plt.plot(x5, y5, "darkred")
plt.show()  # its only shown after the first graph figure tab is closed
```

Exemple: histograms

```python
import matplotlib.pyplot as plt

x6 = [4, 8, 11, 17, 14, 25]
# n = number of elements in each bin, 
# bins = bin limits, 
# patches = list of patches used to create the histogram
n, bins, patches = plt.hist(x6, bins=3)
plt.show()
```

```python
# Histogram with random values

import numpy as np
import matplotlib.pyplot as plt

media = 100
desvio_padrao = 15

valores = media + desvio_padrao * np.random.randn(1000) # 1000 random values
n, bins, patches = plt.hist(valores, 50, edgecolor="black", linewidth=1.0)

plt.xlabel("Valores")
plt.ylabel("Frequência")
plt.title("Histograma")
plt.show()
```

Example: bar graph

```python
import matplotlib.pyplot as plt

pos = [1, 2, 3, 4]
heights = [42, 20, 16, 4]
estados = ["SP", "RJ", "MG", "SC"]
colors = ["darkblue", "darkred", "darkgreen", "darkorange"]

plt.bar(pos, heights, align="center", color=colors)
plt.xticks(pos, estados)
plt.ylabel("Quantidade")
plt.title("Estados Brasileiros")
plt.show()
```

Example: 3D plot

```python
# Hiperbolic paraboloid (horse saddle)
#  z = (-1/6)x**2 + (1/4)y**2
#  x and y are in the range [-20, 10] with 0.5 step

from mpl_toolkits.mplot3d import Axes3D  # 3D graph
from matplotlib import cm  # color map
import matplotlib.pyplot as plt
import numpy as np

fig = plt.figure()
ax = fig.add_subplot(111, projection="3d")  # 1 row, 1 column, 1st figure, 3D graph

x = np.arange(-20, 20.5, 0.5)
y = np.arange(-20, 20.5, 0.5)

# create a matrix with the values of x and y using cartesian product
X, Y = np.meshgrid(x, y)
# Z = (-1/6)*X**2 + (1/4)*Y**2
Z = -(X**2) / 6 + Y**2 / 4

surf = ax.plot_surface(X, Y, Z, cmap=cm.coolwarm, linewidth=0, antialiased=False, rstride=1, cstride=1)
fig.colorbar(surf, shrink=0.5, aspect=5)
plt.show()
```
