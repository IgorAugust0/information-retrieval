# Variáveis em Python

* Variáveis em Python são criadas e destruídas dinamicamente (não se declara variável em Python)
* Variáveis em Python apontam para objetos (são apontadores ou ponteiros)
* Variáveis não possuem tipo (o que posusiu tipo são os objetos apontados pelas variáveis)
* Todo gerenciamento de memória é feito de forma automática em Python.

```python
# isso é comentário em Python
a = 1  # cria a variável "a" e a faz apontar para o objeto 1
```

A atribuição acima fará com que a variável a passe a apontar para o objeto 1.

atribuicao1.jpg

```python
b = a  # essa atribuição faz a variável b apontar para o mesmo objeto que a aponta
```

#Observação:
a atribuição b = a **não gera cópia**. Lembre-se de que as variáveis são ponteiros para objetos. Esse tipo de atribuição faz duas variáveis apontarem para o mesmo objeto. (b apontará para o mesmo objeto apontado por a).

atribuicao2.jpg

Qual é o perigo de duas variáveis apontarem para o mesmo objeto?

* Se o objeto for alterado, as duas variáveis refletirão essa alteração!

Todavia, nesse caso específico (variáveis b e a apontando para o mesmo objeto 1), não corremos esse risco, pois números são objetos **imutáveis** em Python. Significa que se um objeto numérico foi criado com o valor 1, ele não pode mais ser alterado, apenas destruído!
