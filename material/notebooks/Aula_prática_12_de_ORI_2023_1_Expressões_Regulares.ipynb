## Exemplo: obtendo as datas de um texto

Vamos procurar datas nos formatos
* dd/mm/aaaa
* d/mm/aaaa

Vamos dividir em 5 partes:

* Parte 1: os dias dd
[0-9]{1,2}
* Parte 2: a barra /
* Parte 3: os meses mm
[0-9]{1,2}
* Parte 4: a barra /
* Parte 5: o ano (aaaa)
[0-9]{4} - casa com 4 dígitos
([0-9]{2}|[0-9]{4}) - casa 2 ou quatro dígitos

Juntando tudo:

[0-9]{1,2}/[0-9]{1,2}/[0-9]{4}


```python
#obtendo as datas
re.findall( "[0-9]{1,2}/[0-9]{1,2}/[0-9]{4}", texto )
```

A função findall permite que se trabahe com a ideia de subagrupamentos usando parêntestes. Por exemplo, suponha que se deseje trabalhar apenas com os anos
referentes às datas. Podemos usar a mesma expressão regular, mas agora cercamos a perte referente aos meses com um par de parênteses:

```python
#obtendo só os anos das datas cercando a parte do ano com parênteses
re.findall( "[0-9]{1,2}/[0-9]{1,2}/([0-9]{4})", texto )
```

```python
#note que não funcionaria tentar pegar só os 4 dígitos do ano, pois teríamos muito lixo
re.findall( "[0-9]{4}", texto )
```

se você quiser realmente usar o operador parênteses na expressão regular com findall, você eve usar ?: para que o par de parênteses nãos eja tratado como subgrupo

```python
re.findall( "[0-9]{1,2}/[0-9]{1,2}/(?:[0-9]{4})", texto )
```

## Exemplo: obtendo emails da string

Queremos detectar, na string, emaisl como:

* endereco@email.dom
* endereco@email.dom.pr

Paa compor a expressão regular, vamos dividir a tarefa em partes:

* Parte 1: *endereco@*

Supondo que os endereos são compsotos de caracteres minúsculos, dígitos, underline e ponto, seguido de de um @:

[a-z0-9_.]+@

* Parte 2: o domínio *email.*

[a-z0-9_]+

* Parte 3: o domínio geral. Vamos supor que queremos emails com os domínios .com, .org, .net, .edu, .gov

\.(?:com|org|net|edu|gov)

* Parte 4: o domínio de país

(?:\.[a-z]{2}){0,1}

Por fim, compomos nossa expressão regular juntando tudo:

[a-z0-9_.]+@[a-z0-9_]+\.(?:com|org|net|edu|gov)(?:\.[a-z]{2}){0,1}


```python
re.findall( "[a-z0-9_.]+@[a-z0-9_]+.(?:com|org|net|edu|gov)(?:.[a-z]{2}){0,1}", texto )
```
